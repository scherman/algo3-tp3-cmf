```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df4 = pd.read_csv('../../../../datos/tiempos-hconstructiva-mayor-grado-variando-m.csv', sep=', ', engine='python')
# print(df4)
sns.jointplot(df4['m'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'$m$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-hconstructiva-mayor-grado-variando-m.png')
plt.show()
```

```python
df4 = pd.read_csv('../../../../datos/tiempos-hconstructiva-mayor-grado-variando-n.csv', sep=', ', engine='python')
# print(df4)
sns.jointplot(df4['n'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'$n$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-hconstructiva-mayor-grado-variando-n.png')
plt.show()
```

```python
df1 = pd.read_csv('../../../../datos/prom-acierto-hconstructiva-mayor-grado-variando-m.csv', sep=', ', engine='python')
print(df1.columns.tolist())
ax = df1.plot(kind='scatter', x='m', y='promedioAcierto', grid=True, label="Acierto H. Constr.")
ax.set_ylabel("Prom. acierto (%)")
plt.savefig('../../../../pics/prom-acierto-hconstructiva-mayor-grado-variando-m.png')
plt.show()
```

```python
df4 = pd.read_csv('../../../../datos/tiempos-hconstructiva-mayor-grado-variando-n-m0.csv', sep=', ', engine='python')
# print(df4)
sns.jointplot(df4['n'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'$n$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-hconstructiva-mayor-grado-variando-n-m0.png')
plt.show()
```

```python
df4 = pd.read_csv('../../../../datos/tiempos-hconstructiva-mayor-grado-variando-n-mcompleto.csv', sep=', ', engine='python')
# print(df4)
df4['n'] = df4['n'].apply(lambda x: x**2)
sns.jointplot(df4['n'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'$n^2$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-hconstructiva-mayor-grado-variando-n-mcompleto.png')
plt.show()
```

```python
df1 = pd.read_csv('../../../../datos/prom-acierto-hconstructiva-mayor-grado-variando-n-caso-bueno.csv', sep=', ', engine='python')
print(df1.columns.tolist())
ax = df1.plot(kind='scatter', x='n', y='promedioAcierto', grid=True, label="Acierto H. Constr.")
ax.set_ylabel("Prom. acierto (%)")
ax.set_ylim(0,150)
plt.title('Acierto caso bueno')
plt.savefig('../../../../pics/prom-acierto-hconstructiva-mayor-grado-n-caso-bueno.png')
plt.show()
```
