```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random

import math
```

```python
df1 = pd.read_csv('../../../../datos/comp-algoritmos-con-exacto-variando-m.csv', sep=', ', engine='python')
# print(df1.columns.tolist())
dots_size=4
ax1 = df1.plot(kind='scatter', s=dots_size+80, x='m', y='fronteraExacto', grid=True, label="Frontera Exacto")
ax1 = df1.plot(kind='scatter', s=dots_size+5, x='m', y='fronteraHConstructiva', color='g', grid=True, label="Frontera H. Constr.", ax = ax1)
ax1 = df1.plot(kind='scatter', s=dots_size+15, x='m', y='fronteraBLocal', color='r', grid=True, label="Frontera B. Local", ax = ax1)
ax1 = df1.plot(kind='scatter', s=dots_size+5, x='m', y='fronteraGrasp', color='y', grid=True, label="Frontera Grasp", ax = ax1)
plt.title('Grafos tipo random')
plt.ylabel(r'Frontera')
plt.savefig('../../../../pics/fronteras-comp-algoritmos-con-exacto-variando-m.png')
plt.show()
```
