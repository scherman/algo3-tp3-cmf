```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df1 = pd.read_csv('../../../../datos/prom-acierto-blocal-con-heuristica-mayor-grado-variando-m.csv', sep=', ', engine='python')
print(df1.columns.tolist())
ax1 = df1.plot(kind='scatter', x='m', y='promedioAciertoHConstructiva', color='r', grid=True, label="Acierto H. Constr.")
ax2 = df1.plot(kind='scatter', x='m', y='promedioAciertoBLocal', grid=True, label="Acierto B. Local", ax=ax1)
ax2.set_ylabel("Prom. acierto (%)")
plt.savefig('../../../../pics/prom-acierto-blocal-con-heuristica-mayor-grado-variando-m.png')
plt.show()
```

```python
df4 = pd.read_csv('../../../../datos/tiempos-blocal-mayor-grado-variando-n.csv', sep=', ', engine='python')
# print(df5)
sns.jointplot(df4['n'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'$n$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-blocal-mayor-grado-variando-n.png')
plt.show()
```

```python
df4 = pd.read_csv('../../../../datos/tiempos-blocal-mayor-grado-variando-n-mcompleto.csv', sep=', ', engine='python')
# print(df5)
df4['n'] = df4['n'].apply(lambda x: x**2)
sns.jointplot(df4['n'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'$n^2$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-blocal-mayor-grado-variando-n-mcompleto.png')
plt.show()
```
