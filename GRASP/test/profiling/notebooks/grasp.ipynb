```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random

import math
```

```python
import math

df4 = pd.read_csv('../../../../datos/tiempos-grasp-mayor-grado-variando-n.csv', sep=', ', engine='python')
# print(df5)
df4['n'] = df4['n'].apply(lambda x: x * math.log(x,2))
sns.jointplot(df4['n'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'$nlog(n)$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-grasp-mayor-grado-variando-n.png')
plt.show()
```

```python
import math

df4 = pd.read_csv('../../../../datos/tiempos-grasp-mayor-grado-variando-m.csv', sep=', ', engine='python')
# print(df4)
df4['m'] = df4['m'].apply(lambda x: x * math.log(x,2))
sns.jointplot(df4['m'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'$mlog(m)$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-grasp-mayor-grado-variando-m.png')
plt.show()
```

```python
import math

df4 = pd.read_csv('../../../../datos/tiempos-grasp-mayor-grado-variando-n-mcompleto.csv', sep=', ', engine='python')
# print(df5)
df4['n'] = df4['n'].apply(lambda x: (x**2) * math.log(x,2))
sns.jointplot(df4['n'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'${n^2}log(n)$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-grasp-mayor-grado-variando-n-mcompleto.png')
plt.show()
```
