```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
import matplotlib
from itertools import groupby
from operator import itemgetter

import math
```

```python
df4 = pd.read_csv('../../../../datos/tiempos-grasp-mayor-grado-variando-n.csv', sep=', ', engine='python')
# print(df5)
df4['n'] = df4['n'].apply(lambda x: x * math.log(x,2))
sns.jointplot(df4['n'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'$nlog(n)$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-grasp-mayor-grado-variando-n.png')
plt.show()
```

```python
data = pd.DataFrame({
    'tiempo': df2['tiempoTotal'],
    'acierto': df2['promedioAcierto']
})

def example(x):
    return math.floor(data['tiempo'].loc[x] / 1000000.0)

data = data.groupby(by=example).mean()

data['tiempo'] = data['tiempo'].map(lambda x: x / 1000000.0)

ax1 = data.plot(kind='scatter', x='tiempo', y='acierto', grid=True, label="Frontera Exacto")
plt.xlabel('tiempo (ms)')
plt.show()
```
