```python
df3 = pd.read_csv('../../../../datos/tiempos-exacto-variando-n-m0.csv', sep=', ', engine='python')
print(df3.columns.tolist())
df3['n'] = df3['n'].apply(lambda x: x**2)
corr = df3[['n', 'tiempoTotal']].corr()
# PLOTEAR N, TIEMPOTTOTAL CON PEARSON!
ax = df3.plot(kind='scatter', x='n', y='tiempoTotal', title="Pearson: {}".format(corr.iloc[0][1]), grid=True, label="Var. n, $m=0$")
# ax2 = df.plot(kind='scatter', x='c', y='d', color='g', ax=ax1)  
# ax.set_ylim(0, 2*(10**8))
ax.set_ylabel("Tiempo(ns)")
plt.savefig('../../../../pics/tiempos-exacto-variando-n-m0.png')
plt.show()
```

```python
df3 = pd.read_csv('../../../../datos/tiempos-exacto-variando-n-mcompleto.csv', sep=', ', engine='python')
print(df3.columns.tolist())
df3['n'] = df3['n'].apply(lambda x: (2**x)*x)
corr = df3[['n', 'tiempoTotal']].corr()
print (corr)
# print(dir(corr))
# PLOTEAR N, TIEMPOTTOTAL CON PEARSON!
ax = df3.plot(kind='scatter', x='n', y='tiempoTotal', title="Pearson: {}".format(corr.iloc[0][1]),  grid=True, label=r"Var. n, $m=\frac{n(n-1)}{2}$")
# ax2 = df.plot(kind='scatter', x='c', y='d', color='g', ax=ax1)  
# ax.set_ylim(0, 2*(10**8))
ax.set_ylabel("Tiempo(ns)")
plt.savefig('../../../../pics/tiempos-exacto-variando-n-mcompleto.png')
plt.show()
```

```python
from sklearn import datasets, linear_model

df3 = pd.read_csv('../../../../datos/tiempos-exacto-variando-n-m0.csv', sep=', ', engine='python')
print(df3.columns.tolist())
x = df3.n.values
y = df3.tiempoTotal.values

length=1000

x = x.reshape(length, 1)
y = y.reshape(length, 1)

regr = linear_model.LinearRegression()
regr.fit(x, y)

# plot it as in the example at http://scikit-learn.org/
plt.scatter(x, y,  color='black')
plt.plot(x, regr.predict(x), color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()

# df3['n'] = df3['n'].apply(lambda x: x**2)
# corr = df3[['n', 'tiempoTotal']].corr()
# PLOTEAR N, TIEMPOTTOTAL CON PEARSON!
# ax = df3.plot(kind='scatter', x='n', y='tiempoTotal', title="Pearson: {}".format(corr.iloc[0][1]), grid=True, label="Var. n, $m=0$")
# ax2 = df.plot(kind='scatter', x='c', y='d', color='g', ax=ax1)  
# ax.set_ylim(0, 2*(10**8))
# ax.set_ylabel("Tiempo(ns)")
# plt.savefig('../../../../pics/tiempos-exacto-variando-n-m0.png')
# plt.show()
```
