```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df1 = pd.read_csv('../../../../datos/tiempos-exacto-variando-n.csv', sep=', ', engine='python')
# print(df1.columns.tolist())
ax = df1.plot(kind='scatter', x='n', y='tiempoTotal', grid=True, label="Var. n")
ax.set_ylim(0, 0.4*(10**9))
ax.set_ylabel("Tiempo (ns)")
plt.savefig('../../../../pics/tiempos-exacto-variando-n.png')
plt.show()
```

```python
df2 = pd.read_csv('../../../../datos/tiempos-exacto-variando-m.csv', sep=', ', engine='python')
# print(df2.columns.tolist())
ax = df2.plot(kind='scatter', x='m', y='tiempoTotal', grid=True, label="Var. m")
ax.set_ylabel("Tiempo (ns)")
plt.savefig('../../../../pics/tiempos-exacto-variando-m.png')
plt.show()
```

```python

df3 = pd.read_csv('../../../../datos/tiempos-exacto-variando-n-mcompleto.csv', sep=', ', engine='python')
# print(df5)
df3['n'] = df3['n'].apply(lambda x: (2**x)*x)
sns.jointplot(df3['n'], df3['tiempoTotal'], kind="reg")
plt.xlabel(r'${2^n}n$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-exacto-variando-n-mcompleto.png')
plt.show()
```

```python
df4 = pd.read_csv('../../../../datos/tiempos-exacto-variando-n-m0.csv', sep=', ', engine='python')
# print(df5)
df4['n'] = df4['n'].apply(lambda x: x**2)
sns.jointplot(df4['n'], df4['tiempoTotal'], kind="reg")
plt.xlabel(r'$n^2$')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('../../../../pics/tiempos-exacto-variando-n-m0.png')
plt.show()
```
